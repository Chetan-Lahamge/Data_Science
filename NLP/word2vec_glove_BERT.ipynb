### BERT - Glove

```python
from flair.embeddings import WordEmbeddings, TransformerWordEmbeddings
from flair.data import Sentence
from scipy.spatial import distance
```

```python
glove_embedding = WordEmbeddings('glove')
```

```python
sentence = Sentence("The founder of Apple is Steve Jobs")
```

```python
glove_embedding.embed(sentence)
```

```python
for token in sentence:
    print(token)
    print(token.embedding)
    print('-----------------------------')
```

```python
sentence_2 = Sentence("Apple is a Fruit")
glove_embedding.embed(sentence_2)

for token in sentence_2:
    print(token)
    print(token.embedding)
    print("-------------------------")
```

```python
bert_embeddings = TransformerWordEmbeddings('bert-base-multilingual-cased')
```

```python
bert_embeddings.embed(sentence)
```

```python
for token in sentence:
    print(token)
    print(token.embedding)
    print('-----------------------')
```

```python
bert_embeddings.embed(sentence_2)
```

```python
for token in sentence_2:
    print(token)
    print(token.embedding)
    print('-----------------------')
```

```python
import numpy as np
```

```python
distance.euclidean(np.array(sentence[3].embedding),np.array(sentence_2[0].embedding)) ### GloVe Embedding Distance
```
