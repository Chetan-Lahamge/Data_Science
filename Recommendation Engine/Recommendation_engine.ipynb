### 1. Correlation Matrix for Recommendations

```python
movies_df["userId"].unique()
```

```python
user_recomm_corr = pd.pivot_table(data=movies_df,values="rating",index="movie",columns="userId").fillna(0)
user_recomm_corr.head()
```

```python
user_recomm_corr.columns = movies_df['userId'].unique()
```

```python
user_recomm_corr
```

### 2. Eclideon Distance metric For Recommendations

```python
from sklearn.metrics import pairwise_distances
```

```python
user_recomm_transpose = user_recomm_corr.T
```

```python
user_recomm_euclideon = user_recomm_transpose.values
```

```python
eclidean_distance_df = pairwise_distances(X=user_recomm_euclideon,metric='euclidean')
```

```python
euclidean_distance_df_new = pd.DataFrame(data=eclidean_distance_df,columns=movies_df['userId'].unique(),index=movies_df['userId'].unique())
```

```python
euclidean_distance_df_new.iloc[:15,:15].idxmin()
```

```python
euclidean_distance_df_new.iloc[:15,:15]
```

```python
import numpy as np
```

```python
np.fill_diagonal(euclidean_distance_df_new.to_numpy(),val=100)
```

```python
euclidean_distance_df_new.iloc[:15,:15].idxmin()
```

### Cosine Similarity for Recommendation Engine

```python
cosine_sim_df = 1- pairwise_distances(X=user_recomm_euclideon,metric='cosine')
cosine_sim_df_new = pd.DataFrame(data=cosine_sim_df,columns=movies_df['userId'].unique(),index=movies_df['userId'].unique())
```

```python
np.fill_diagonal(cosine_sim_df_new.to_numpy(),val=0)
```
